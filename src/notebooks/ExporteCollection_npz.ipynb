```python
from concurrent.futures import ThreadPoolExecutor, as_completed
import time

PIXEL_SIZE = 30
CHIP_SIZE = 256

@retry(tries=2, delay=5, backoff=3)
def downloadEachImage(args):
    """Download each file of list."""
    expression = args[0]
    region_dict = args[1]
    name = args[2]

    region = ee.Geometry(region_dict)
    proj = ee.Projection("EPSG:4326").atScale(PIXEL_SIZE).getInfo()
    scale_y = -proj["transform"][0]
    scale_x = proj["transform"][4]

    listCoords = ee.Array.cat(region.coordinates(), 1).getInfo()
    centroid = region.centroid(0.001).coordinates().getInfo()
    lon, lat = centroid[0], centroid[1]
    xMin = listCoords[0][0]
    yMax = listCoords[2][1]
    coords = [xMin, yMax]

    grid = {
        "dimensions": {"width": CHIP_SIZE, "height": CHIP_SIZE},
        "affineTransform": {
            "scaleX": scale_x,
            "scaleY": scale_y,
            "translateX": coords[0],
            "translateY": coords[1],
        },
        "crsCode": "EPSG:4326",
    }

    response = session.post(
        url=url,
        data=json.dumps(
            {
                "expression": ee.serializer.encode(expression),
                "fileFormat": "NPY",
                "bandIds": expression.bandNames().getInfo(),
                "grid": grid,
            }
        ),
    )

    try:
        error = json.loads(response.content)
        logger = logging.getLogger()
        logger.error(str(json.loads(response.content)))
        raise Exception("run again")
    except UnicodeDecodeError:
        array = np.load(io.BytesIO(response.content)).tolist()
        array = np.array(array)

        np.savez_compressed(
            f"/content/drive/MyDrive/ClayModel_detail/palm_ls/data/binaryclass/raw_test/{name}.npz",
            #f"/content/{name}.npz",
            array=array.transpose([2, 0, 1]),
            timestamp=expression.get('timestamp').getInfo(),
            lon=expression.get('longitude').getInfo(),
            lat=expression.get('latitude').getInfo(),
            gsd=expression.get('gsd').getInfo(),
        )

def parallel_download(tasks):
    """
    Execute downloadEachImage in parallel.

    Args:
        tasks: List of arguments for downloadEachImage.
        max_workers: Number of threads to use for parallel execution.
    """
    start_time = time.time()

    with ThreadPoolExecutor() as executor:
        future_to_task = {executor.submit(downloadEachImage, task): task for task in tasks}

        for future in as_completed(future_to_task):
            task = future_to_task[future]
            try:
                future.result()
                print(f"Task completed: {task[2]}")
            except Exception as exc:
                print(f"Task {task[2]} generated an exception: {exc}")

    end_time = time.time()
    print(f"Total execution time: {(end_time - start_time) / 60:.2f} minutes")
```

```python
import datetime

def mask_ls_clouds(image):
    """
    Aplica uma máscara para remover nuvens, sombras, neve, água e cirros de alta confiança com base na banda BQA.

    Args:
        image (ee.Image): Landsat com a banda 'BQA'.

    Returns:
        ee.Image: Imagem com máscara aplicada.
    """
    # Seleciona a banda BQA
    qa = image.select('QA_PIXEL')

    # Cria a máscara com condições baseadas em bits da banda BQA
    mask = (
        qa.bitwiseAnd(1 << 3)  # Cloud com qualquer nível de confiança
        .And(qa.bitwiseAnd(1 << 8).Or(qa.bitwiseAnd(1 << 9)))
        .Or(qa.bitwiseAnd(1 << 1))  # Dilated cloud
        .Or(qa.bitwiseAnd(1 << 4).And(qa.bitwiseAnd(1 << 10).Or(qa.bitwiseAnd(1 << 11))))  # Cloud shadow
        .Or(qa.bitwiseAnd(1 << 5))  # Snow
        .Or(qa.bitwiseAnd(1 << 7))  # Water
        .Or(qa.bitwiseAnd(1 << 14).And(qa.bitwiseAnd(1 << 15)))  # Cirrus (alta confiança)
    )
    #mask = (
    #qa.bitwiseAnd(1 << 3)  # Cloud com qualquer nível de confiança
    #.And(qa.bitwiseAnd(1 << 8))  # Somente nuvens de alta confiança
    #.Or(qa.bitwiseAnd(1 << 1))  # Dilated cloud
    #)

    # Retorna a imagem com a máscara aplicada
    return image.updateMask(mask.Not())

def renameBands(image):
    return image.rename(['red', 'green', 'blue', 'nir08', 'swir16', 'swir22'])

# Criar uma grade regular
def regular_grid(geometry, edge_size, centroid):
    grid = geometry.coveringGrid(ee.Image().projection(), edge_size)
    return ee.FeatureCollection(ee.Algorithms.If(
        centroid,
        grid.map(lambda ft: ft.setGeometry(ft.geometry().centroid(0.001))),
        grid
    ))

l9l8_bands = ['SR_B4', 'SR_B3', 'SR_B2', 'SR_B5', 'SR_B6', 'SR_B7'];
l7l5_bands = ['SR_B3', 'SR_B2', 'SR_B1', 'SR_B4', 'SR_B5', 'SR_B7'];

year = 2012

# Parâmetros de análise
start_date = f"{year}-01-01"
end_date = f"{year}-12-31"

end_date_obj = datetime.date.fromisoformat(end_date)

landsat7_end = datetime.date(2003, 4, 30)

print(f"Start date: {landsat7_end}")
print(f"End date: {end_date}")

# Inicializa uma coleção vazia
dataset_landsat = ee.ImageCollection([])

# Carrega a coleção Landsat Collection 2 tier 1
landsat9 = (ee.ImageCollection('LANDSAT/LC09/C02/T1_L2')
                    .filterDate(start_date, end_date)
                    .filterBounds(cartas)
                    .filter(ee.Filter.lt('CLOUD_COVER_LAND', 80))
                    .map(mask_ls_clouds)
                    .select(l9l8_bands)
                    .map(renameBands)
                    )
dataset_landsat = dataset_landsat.merge(landsat9)

# Carrega a coleção Landsat Collection 2 tier 1
landsat8 = (ee.ImageCollection('LANDSAT/LC08/C02/T1_L2')
                    .filterDate(start_date, end_date)
                    .filterBounds(cartas)
                    .filter(ee.Filter.lt('CLOUD_COVER_LAND', 80))
                    .map(mask_ls_clouds)
                    .select(l9l8_bands)
                    .map(renameBands)
                    )
dataset_landsat = dataset_landsat.merge(landsat8)

# # Carrega Landsat 7 se end_date for antes do seu desligamento
# if end_date_obj <= landsat7_end:
#     landsat7 = (ee.ImageCollection('LANDSAT/LE07/C02/T1_L2')
#                         .filterDate(start_date, end_date)
#                         .filterBounds(cartas)
#                         .filter(ee.Filter.lt('CLOUD_COVER_LAND', 80))
#                         .map(mask_ls_clouds)
#                         .select(l7l5_bands)
#                         .map(renameBands))
#     dataset_landsat = dataset_landsat.merge(landsat7)
# else:
#   landsat7=0
#   print("Sem coleção 7...")

# Carrega Landsat 7 se end_date for antes do seu desligamento
landsat7 = (ee.ImageCollection('LANDSAT/LE07/C02/T1_L2')
                    .filterDate(start_date, end_date)
                    .filterBounds(cartas)
                    .filter(ee.Filter.lt('CLOUD_COVER_LAND', 80))
                    .map(mask_ls_clouds)
                    .select(l7l5_bands)
                    .map(renameBands))
dataset_landsat = dataset_landsat.merge(landsat7)


#
# Carrega a coleção Landsat Collection 2 tier 1
landsat5 = (ee.ImageCollection('LANDSAT/LT05/C02/T1_L2')
                    .filterDate(start_date, end_date)
                    .filterBounds(cartas)
                    .filter(ee.Filter.lt('CLOUD_COVER_LAND', 100))
                    .map(mask_ls_clouds)
                    .select(l7l5_bands)
                    .map(renameBands))
dataset_landsat = dataset_landsat.merge(landsat5)

# Ordena as imagens pelo atributo 'CLOUD_COVER'
dataset_landsat = dataset_landsat.sort('CLOUD_COVER')

# Criar uma grade regular
r_grid = regular_grid(cartas.geometry(), 512 * 30, False)

# def filter_mostly_contained(collection, roi):

#     def map_containment(image):
#         footprint = image.geometry()
#         roi_geometry = roi.geometry()
#         intersection = footprint.intersection(roi_geometry, ee.ErrorMargin(1))
#         intersection_area = intersection.area(ee.ErrorMargin(1))
#         roi_area = roi_geometry.area(ee.ErrorMargin(1))
#         containment_fraction = intersection_area.divide(roi_area)

#         return image.set("containment_fraction", containment_fraction)

#     collection_with_containment = collection.map(map_containment)
#     max_containment = collection_with_containment.aggregate_max("containment_fraction")

#     return collection_with_containment.filter(ee.Filter.eq("containment_fraction", max_containment))

def filter_most_valid_pixels(collection, roi):
    """
    Filters an ImageCollection to return images with the highest number of valid (non-masked) pixels inside the ROI.
    The output is sorted in descending order of valid pixel count.

    Args:
        collection (ee.ImageCollection): The input image collection.
        roi (ee.Geometry): The region of interest.

    Returns:
        ee.ImageCollection: The filtered collection sorted by valid pixel count.
    """
    def compute_valid_pixels(image):
        mask = image.mask()  # Get the image mask (valid pixels = 1, masked = 0)
        valid_pixels = mask.reduceRegion(
            reducer=ee.Reducer.sum(),  # Count non-masked pixels
            geometry=roi,
            scale=30,  # Adjust scale based on dataset resolution
            bestEffort=True
        ).get("green")  # Use any band (assuming all bands have the same mask)

        return image.set("valid_pixel_count", valid_pixels)

    # Compute valid pixel count for each image
    collection_with_counts = collection.map(compute_valid_pixels)

    # Sort by valid pixel count in descending order
    return collection_with_counts.sort("valid_pixel_count", False)


# Função para selecionar a imagem com menor índice de nuvens para cada tile
def filterImageByTiles(tile):
    geometry = tile.geometry()
    tile_id = tile.get("system:index")

    ## Filtrar a coleção para imagens que intersectam o tile
    #filtered_images = dataset_landsat.filterBounds(geometry.centroid(0.001))
    #imageByTile = ee.ImageCollection(filtered_images).first().reproject(crs=filtered_images.first().select('green').projection(),scale=filtered_images.first().select('green').projection().nominalScale())


    # Filtrar a coleção para imagens que intersectam o tile
    filtered_images = dataset_landsat.filterBounds(geometry.centroid(0.001))
    images_mostly_contained = filter_most_valid_pixels(filtered_images, tile)
    imageByTile = ee.ImageCollection(images_mostly_contained).first().reproject(crs=images_mostly_contained.first().select('green').projection(),scale=images_mostly_contained.first().select('green').projection().nominalScale())


    # Adicionar propriedades do tile e do ROI
    if imageByTile:
        clipped_image = imageByTile.clip(geometry).set({
            "tile_id": tile_id,
            "timestamp": ee.Date(filtered_images.first().get('system:time_start')).format('yyyy-MM-dd HH:mm:ss.SSSSSSZZZZ'),
            "latitude": geometry.centroid(0.001).coordinates().get(1),
            "longitude": geometry.centroid(0.001).coordinates().get(0),
            "gsd": imageByTile.get('resolution_meters'),
            "system:footprint": tile.geometry(),
            "grid_index": tile.get('system:index'),
        }).set(filtered_images.first().toDictionary())
        #print(clipped_image.getInfo())
        return clipped_image
    else:
        return None

grid_ids = r_grid.aggregate_array('system:index').getInfo()

tiles_with_images = []

for grid_id in grid_ids:  # Iterate through each grid_id in the list
    x = grid_id
    grid = r_grid.filter(ee.Filter.eq('system:index', x))
    #print(grid.size().getInfo())
    tile_with_image = ee.Image(filterImageByTiles(grid))
    tiles_with_images.append(tile_with_image)

# Criar uma coleção de imagens representativas
tiles_collection = ee.ImageCollection(tiles_with_images)
```

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
import time
import rasterio
import numpy as np
import matplotlib.pyplot as plt

PIXEL_SIZE = 30
CHIP_SIZE = 512
TARGET = "dende"
REGION = 'BA'
SENSOR = "landsat"
YEAR = "2012"

def plot_rgb_tiff(tiff_path):

    with rasterio.open(tiff_path) as src:
        print(src)
        # Lê as bandas 2, 1, 0 (considerando indexação 0)
        red = src.read(1)  # Banda 2
        green = src.read(2)  # Banda 1
        blue = src.read(3)  # Banda 0

        # Normaliza os valores para [0, 1] para visualização
        def normalize(band):
            band_min, band_max = band.min(), band.max()
            return (band - band_min) / (band_max - band_min + 1e-6)

        red_norm = red*2.75e-05
        green_norm = green*2.75e-05
        blue_norm = blue*2.75e-05

        # Empilha as bandas para formar a imagem RGB
        rgb = np.dstack((red_norm, green_norm, blue_norm))

        # Plotagem
        plt.figure(figsize=(10, 10))
        plt.imshow(rgb)
        plt.axis('off')
        plt.title('Imagem RGB (Bandas 2, 1, 0)')
        plt.show()


# Função de download para cada imagem
@retry(tries=2, delay=5, backoff=2)
def downloadEachImage(args):
    """Download each file of list."""
    expression = args[0]
    region_dict = args[1]
    name = args[2]

    region = ee.Geometry(region_dict)
    proj = ee.Projection("EPSG:4326").atScale(PIXEL_SIZE).getInfo()
    scale_y = -proj["transform"][0]
    scale_x = proj["transform"][4]

    listCoords = ee.Array.cat(region.coordinates(), 1).getInfo()
    centroid = region.centroid(0.001).coordinates().getInfo()
    lon, lat = centroid[0], centroid[1]
    xMin = listCoords[0][0]
    yMax = listCoords[2][1]
    coords = [xMin, yMax]

    grid = {
        "dimensions": {"width": CHIP_SIZE, "height": CHIP_SIZE},
        "affineTransform": {
            "scaleX": scale_x,
            "scaleY": scale_y,
            "translateX": coords[0],
            "translateY": coords[1],
        },
        "crsCode": "EPSG:4326",
    }

    response = session.post(
        url=url,
        data=json.dumps(
            {
                "expression": ee.serializer.encode(expression),
                "fileFormat": "GEO_TIFF",
                "bandIds": expression.bandNames().getInfo(),
                "grid": grid,
            }
        ),
    )

    try:
        error = json.loads(response.content)
        logger = logging.getLogger()
        logger.error(str(json.loads(response.content)))
        raise Exception("run again")
    except UnicodeDecodeError:
        # Caminho de saída para salvar o GeoTIFF
        geotiff_path = f"/content/drive/MyDrive/ClayModel_detail/palm_ls/data/predict/{TARGET}/{YEAR}/{REGION}/{name}.tif"

        # Salva o conteúdo da resposta diretamente como GeoTIFF
        with open(geotiff_path, "wb") as f:
            f.write(response.content)

        print(f"Arquivo GeoTIFF salvo em: {geotiff_path}")


def parallel_download(tasks):
    """
    Realiza o download das imagens em paralelo.

    Args:
        tasks: Lista de argumentos para a função downloadEachImage.
        max_workers: Número de threads para execução paralela.
    """
    with ThreadPoolExecutor() as executor:
        futures = {executor.submit(downloadEachImage, task): task for task in tasks}

        for future in as_completed(futures):
            task = futures[future]
            try:
                future.result()
                print(f"Download concluído para: {task[2]}")
            except Exception as exc:
                print(f"Erro no download para {task[2]}: {exc}")


def process_and_download_in_chunks(tiles_list, chunk_size=10):
    """
    Processa uma lista de imagens em blocos e baixa cada bloco imediatamente.

    Args:
        tiles_list: Lista de imagens do Earth Engine.
        chunk_size: Número de imagens por bloco a processar.
    """
    print(f"Número total de imagens: {len(tiles_list)}")

    for start in range(0, len(tiles_list), chunk_size):
        end = min(start + chunk_size, len(tiles_list))
        print(f"Processando imagens de {start} a {end - 1}")

        start_time = time.time()

        # Criar argumentos para o bloco atual
        tasks = []
        for i in range(start, end):
            image = tiles_list[i]  # Acessar diretamente a imagem na lista
            try:
                region_dict = image.get('system:footprint').getInfo()
            except Exception as e:
                print(f"Erro ao obter a geometria para a imagem {i}: {e}")
                continue  # Ignorar a imagem atual e continuar

            name = f'{SENSOR}{YEAR}_{i}'
            tasks.append((image, region_dict, name))

        # Fazer o download das tarefas do bloco atual
        if tasks:  # Apenas realizar download se houver tarefas válidas
            parallel_download(tasks)

        end_time = time.time()
        duration = (end_time - start_time) / 60
        print(f"Duração do processamento do bloco {start}–{end - 1}: {duration:.2f} min")

    print("Processamento concluído.")
```
